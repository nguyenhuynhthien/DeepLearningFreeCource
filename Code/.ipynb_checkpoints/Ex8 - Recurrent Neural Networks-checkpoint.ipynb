# Recurrent Neural Networks

## Import modules

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow.keras.backend as K
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import LSTM
```

```python
train = pd.read_csv('../data/Google_Stock_Price_Train.csv')
train['Date'] = pd.to_datetime(train['Date'])
train = train.set_index('Date')
train.head()
```

```python
train.info()
```

```python
train = train.loc[:, ['Close']] #.apply(lambda x : x.replace(',', '')).astype('float32')
train = train.replace(',','', regex=True)
train['Close'] = train['Close'].astype('float32')
train.head(20)
```

```python
train.plot()
```

```python
test = pd.read_csv('../data/Google_Stock_Price_Test.csv')
test['Date'] = pd.to_datetime(test['Date'])
test = test.set_index('Date')
test.head()
```

```python
test = test.loc[:, ['Close']]
test.head()
# test.plot()
```

```python
ax = train.plot()
test.plot(ax=ax)
plt.legend(['train', 'test'])
```

## Scaling, tch train-test

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
train_sc = sc.fit_transform(train)
test_sc = sc.transform(test)
```

```python
train_sc[:4]
```

```python
X_train = train_sc[:-1]
y_train = train_sc[1:]

X_test = test_sc[:-1]
y_test = test_sc[1:]
```

```python
early_stop = EarlyStopping(monitor='loss', patience=3, verbose=1)
```

```python
train_sc.shape
```

```python
train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
train_sc_df.head()
```

## To cc rolling window c kch thc 20 ngy

```python
window_size = 5 # Ch  chn window_size sao cho khng  mt d liu test vn c t data point
for s in range(1, window_size+1):
    train_sc_df['{} days before'.format(s)] = train_sc_df['Scaled'].shift(s)
    test_sc_df['{} days before'.format(s)] = test_sc_df['Scaled'].shift(s)
```

```python
train_sc_df.head(window_size)
```

```python
X_train = train_sc_df.dropna().drop('Scaled', axis=1)
y_train = train_sc_df.dropna()[['Scaled']]

X_test = test_sc_df.dropna().drop('Scaled', axis=1)
y_test = test_sc_df.dropna()[['Scaled']]
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train = X_train.values
X_test= X_test.values

y_train = y_train.values
y_test = y_test.values
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 1, window_size)
X_test_t = X_test.reshape(X_test.shape[0], 1, window_size)
```

```python
X_test_t.shape
```

### To model RNN

```python
K.clear_session()
model = Sequential()
# Thm lp LSTM, thit lp return_sequences=True  mi node ra tr ra mt output
model.add(LSTM(20, input_shape=(1, window_size), return_sequences=True))
model.add(LSTM(20))
# Thm lp DNN u ra
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1))
# Thit lp thng s
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
model.fit(X_train_t, y_train, epochs=100,
          batch_size=3, verbose=1, callbacks=[early_stop])
```

```python
y_pred = model.predict(X_test_t)
plt.plot(y_test)
plt.plot(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred, squared=False)
```
