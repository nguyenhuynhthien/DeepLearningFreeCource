# Import module

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([4, 9, 6, 3, 8])
y2 = np.array([8, 3, 8, 6, 1])
```

# Plot

```python
plt.plot(X, y)
```

## Label, title

```python
plt.plot(X, y)
plt.title('My plot')
plt.xlabel('X value')
plt.ylabel('y value')
```

## Color

```python
plt.plot(X, y, c='r')
plt.title('My plot')
plt.xlabel('X value')
plt.ylabel('y value')
```

# Vẽ nhiều đồ thị trong 1 hình

```python
plt.plot(X, y, X, y2)
plt.title('My plot')
plt.xlabel('X value')
plt.ylabel('y value')
```

# Vẽ nhiều đồ thị

```python
plt.subplot(1, 2, 1)
plt.plot(X, y)
plt.xlabel('X value')
plt.ylabel('y value')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot(X, y2)
plt.xlabel('X value')
plt.ylabel('y value')
plt.title('Plot 2')

plt.suptitle('Two plot in a figure')
```

## Xem thêm hàm subplots()

# Bar

```python
X2 = ['Value 1', 'Value 2', 'Value 3', 'Value 4', 'Value 5']
```

```python
plt.bar(X2, y2)
```

## Legend

```python
plt.bar(X2, y2)
plt.xlabel('X2 value')
plt.ylabel('y2 value')
plt.title('Bar graph')
plt.legend(['Parameters'])
```
