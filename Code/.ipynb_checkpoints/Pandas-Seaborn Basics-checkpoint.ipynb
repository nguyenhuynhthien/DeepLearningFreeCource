# Vẽ đồ thị

```python
df.plot()
```

```python
df['age'].plot()
```

```python
df.plot(kind = 'scatter', 
        x = 'age', 
        y = 'price', 
        xlabel = 'Age', 
        ylabel = 'Price', 
        title = 'Price vs Age')
```

```python
df['price'].plot(kind = 'hist', bins = 11, xlabel = 'Price', title = 'Histogram of Price')
```

```python
import seaborn as sns
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
sns.pairplot(df, hue = 'is new')
```
