# Multiclass classification/ Phân loại nhiều lớp với DNN

## Import motdules

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import EarlyStopping
```

## Load dataset

```python
df = pd.read_csv('../data/fetal_health.csv')
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop('fetal_health', axis=1)
X.head()
```

## Mã hoá lại y

```python
target_names = df['fetal_health'].unique()
target_names
```

```python
target_dict = {n:i for i, n in enumerate(target_names)}
target_dict
```

```python
y= df['fetal_health'].map(target_dict)
y.head(10)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_cat = to_categorical(y, dtype = int)
```

```python
y_cat[:10]
```

### Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y_cat,
                                                    test_size=0.2,
                                                   random_state=10)
```

## Tạo model

```python
model = Sequential()
model.add(Dense(64, input_shape=(21,), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
# model.add(Dense(100, activation='relu'))
# model.add(Dense(12, activation='relu'))
# model.add(Dense(10, activation='relu'))
# model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=150, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:5]
```

```python
y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_class, y_pred_class))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(y_test_class, y_pred_class)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test_class, y_pred)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss: " + str(loss));
print("Accuracy: " + str(accuracy));
```
